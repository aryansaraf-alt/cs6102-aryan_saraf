```python
#------------------------------ENTER YOUR CODE HERE--------------------------------------------#
#defining trace region and other important variables
pt = textin_array
sbox_ = np.array(sbox)

num_traces = trace_array.shape[0]
num_points = trace_array.shape[1]

region_start = 2000
region_end = 3000
traces = trace_array[:, region_start:region_end]

traces_centered = traces - np.mean(traces, axis=0)
traces_std = np.std(traces, axis=0)

key = []

for idx in range(16):
    #hamming weight precomputation
    hw = np.array([bin(n).count('1') for n in range(256)], dtype=np.float64)

    #hypothesis matrix
    hypothesis = hw[sbox_[pt[:, idx][:, None] ^np.arange(256)]]

    #centering the hypothesis
    hypothesis_centered = hypothesis - np.mean(hypothesis, axis=0)
    hypothesis_std = np.std(hypothesis, axis=0)

    #pearsons correlation
    corr = hypothesis_centered.T @ traces_centered #T -> transpose and @-> matrix multiplication
    corr /= (num_traces *hypothesis_std[:, None]*traces_std[None, :])

    max_corr = np.max(np.abs(corr), axis=1)
    best_key = np.argmax(max_corr)

    key.append(best_key)
    print(f"byte {idx}: {best_key}")


```

```python
#------------------------------DO NOT EDIT THE BELOW CODE--------------------------------------------#
print("Key guess: ", np.array(key))
print("actual key:", key_array[0])
if np.array_equal(np.array(key), key_array[0]):
    print("Keys match!")
else:
    print("Keys do not match.")
```
